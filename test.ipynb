```python
def str_to_datetime(string):
    if not string:
        return None
    string = string.replace('T', '-')
    string = string.replace(':', '-')

    y, mon, d, h, min, s = string.split('-')
    date_time_str = d + '/' + mon + '/' + y[-2:] + ' ' + h + ':' + min + ':' + s
    
    return datetime.strptime(date_time_str, '%d/%m/%y %H:%M:%S')
```
